### 设定参数

输入文本指令

````python
# text='white hair'  
````

输入图片，请将图片放在input_img/文件夹下

````python
# input_img='input_img/img3.png'  
````

指定修改区域，可以是:
1. None，即整幅图微调，如“smiling”。 
2. 一块矩形区域，输入一个字典，```{‘bbox’: bounding box}```，bounding box是一个列表，分别代表矩形的[上、下、左、右]。
3. 指定部位，输入一个字典，```{‘organ’: organ_name}```，oragan_name是一个列表，可以在```[ 'skin','nose','eye_g','l_eye', 'r_eye','l_brow', 'r_brow','l_ear','r_ear', 'mouth', 'u_lip', 'l_lip', 'hair', 'hat','ear_r'，'neck_l', 'neck','cloth']```中选择，其中eye_g为眼睛，l,r表示左右，'u_lip', 'l_lip'表示上下唇。ear_r为耳坠，neck_l为项链。

````python
# region={'bbox':[232,700,52,605]}
````

可以指定一些训练超参数，更多的可以参考config文件的Option类，否则采用默认值。

````python
text='Mohawk'  
input_img='input_img/img6.png' 
region=None
opts = Options().get_args()
opts.latent_lambda = 0.0012
opts.img_lambda = 0
opts.id_lambda = 0
opts.step = 30
````

### 开始优化

````python
result = optim(text=text, input_img=input_img, opts=opts, region=region)
````

### 结果展示

````python
%matplotlib inline

from torchvision.utils import make_grid
from torchvision.transforms import ToPILImage
result_image = ToPILImage()(make_grid(result.detach().cpu(), normalize=True, scale_each=True, range=(-1, 1), padding=0))
h, w = result_image.size
result_image = result_image.resize((h // 2, w // 2))
# result_image
import matplotlib.pyplot as plt
plt.figure(dpi=200)
# plt.figure(figsize=(6, 100))
plt.axis('off')
plt.imshow(result_image)
plt.title('input command: {}'.format(text),fontsize =7)
plt.text(200,600,'origin',fontsize =7)
plt.text(700,600,'invert',fontsize =7)
plt.text(1200,600,'edit',fontsize =7)
if region:
    plt.text(1700,600,'mask',fontsize =7)
plt.show()
````
